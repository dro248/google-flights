Google Flights URL breakdown  
==========================  
  
Part 1 - Host
------------------
**host:** https://www.google.com/flights  
?  
**requested language:** hl=en  
\#  

------
  
  
Part 2 - Trip Type
-------------------------
###### Round trip
flt=/m/0f2r6.__[departure airport]__.__[departure date]__\*__[return airport]__./m/0f2r6.__[return date]__  
  
  
###### One Way
TODO

Part 3 - Key Value Pairs
----------------------------------
  
##### Currency  
c:USD  

------
##### Number of Stops  
s:{'any number of stops': omit, '1 or fewer': 1, '2 or fewer': 2}  

------
##### _____ (necessary?)
e:1  
  
------
###### Max Flight Duration *(optional)*
md:[in_flight_duration_minutes]__\*__[in_flight_duration_minutes]  
  
**example:**  
max_inflight_duration = 10hours (or 300 minutes)
  
`md:300*300`  

------
##### Max Price per Ticket  *(optional)*
p: number_of_cents  
  
**example:**  
max_price = $5400.00  
  
`p:540000`

------
##### Bags
b: carry-on, checked  

------
##### Passengers
px:adults  

------
##### airline
a:AA\*AA  

------ 
##### Seat Class  
* 'economy': omit 
* 'premium economy': p
* 'business': b 
* 'first class': f
  
**example:**    
`sc:b`  

------
##### ???  
sd:1  

------
##### ???  
t:f  

```python
import requests
from bs4 import BeautifulSoup
```

### Example URL

```python
url = 'https://www.google.com/flights?hl=en#flt=/m/0f2r6.MIA.2019-02-13*MIA./m/0f2r6.2019-02-17;c:USD;e:1;a:AA*AA;sd:1;t:f'
```

```python
page = requests.get(url).text
page
```

```python
soup = BeautifulSoup(page)
soup
```

```python
print(soup.prettify())
```
